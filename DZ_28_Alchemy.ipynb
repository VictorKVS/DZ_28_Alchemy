### **Домашнее задание**

#### **Задача 1: Оптимизация загрузки данных**  
1. Создайте две связанные модели: `Author` и `Book` с отношением **один ко многим**.  
   - `Author` содержит поля: `id`, `name`.  
   - `Book` содержит поля: `id`, `title`, `author_id` (внешний ключ).  

2. Напишите два запроса:  
   - Первый с использованием **ленивой загрузки** (Lazy).  
   - Второй с использованием **жадной загрузки** (Eager, `joinedload`).  

3. Сравните количество выполняемых SQL-запросов в обоих случаях и объясните разницу. 

#### **Задача 2: Транзакции и откаты изменений**  
1. Создайте таблицу `User` с полями: `id`, `username`, `email`.  
2. Напишите код, который выполняет следующие действия в рамках транзакции:  
   - Добавляет два новых пользователя.  
   - Имитация ошибки при добавлении третьего пользователя.  

3. Реализуйте механизм, который **откатит изменения** в случае возникновения ошибки и оставит базу данных в неизменном состоянии.  
4. Проверьте, что данные откатываются корректно.  

#### **Задача 3: Работа с Alembic**  
1. **Инициализируйте Alembic** в вашем проекте (если еще не выполнено).  
2. Создайте и примените миграции для добавления новой таблицы `Order` с полями:  
   - `id` (первичный ключ),  
   - `product_name` (строка),  
   - `quantity` (целое число),  
   - `created_at` (дата и время).  

3. Внесите изменения в структуру таблицы `Order`:  
   - Добавьте поле `price` (с плавающей точкой).  
   - Удалите поле `created_at`.  

4. Откатите последнюю миграцию и убедитесь, что структура таблицы вернулась к предыдущему состоянию.  

#### **Задача 4: Паттерн Repository**  
1. Реализуйте паттерн **Repository** для модели `Book`, который будет включать следующие методы:  
   - Добавление новой книги.  
   - Получение всех книг автора по его `id`.  
   - Удаление книги по её `id`.  

2. Создайте фабрику сессий для управления сессиями SQLAlchemy.  
3. Напишите несколько тестовых сценариев для проверки работы методов репозитория.
